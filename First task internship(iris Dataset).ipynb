Visualization


Scatter Plot:-

```python
from itertools import combinations

feature_cols = ["sepal_length", "sepal_width",
                "petal_length", "petal_width"]

# 5) Loop over every pair and plot
for x_feat, y_feat in combinations(feature_cols, 2):
    plt.figure(figsize=(6, 4))

    # ──► If you have a 'species' column, colour by species
    if "species" in dataset.columns:
        for sp, grp in dataset.groupby("species"):
            plt.scatter(grp[x_feat], grp[y_feat],
                        label=sp, alpha=0.75,
                        edgecolor="k", linewidth=0.4)
        plt.legend(title="Species")
    else:
        plt.scatter(dataset[x_feat], dataset[y_feat],
                    alpha=0.75, edgecolor="k", linewidth=0.4)

    # Labels, title
    plt.xlabel(x_feat.replace("_", " ").title())
    plt.ylabel(y_feat.replace("_", " ").title())
    plt.title(f"{x_feat} vs {y_feat}")
    plt.tight_layout()
    plt.show()
```

Boxplots to identify outliers

```python
plt.figure(figsize=(4, 6))
plt.boxplot(dataset["sepal_length"],
            vert=True, patch_artist=True,
            boxprops=dict(facecolor="lightblue", edgecolor="k"),
            medianprops=dict(color="red"),
            whiskerprops=dict(color="k"),
            capprops=dict(color="k"),
            flierprops=dict(marker="o", markerfacecolor="orange",
                            markersize=6, markeredgecolor="k"))
plt.title("Boxplot of Sepal Length")
plt.ylabel("Sepal length (cm)")
plt.tight_layout()
plt.show()
```
